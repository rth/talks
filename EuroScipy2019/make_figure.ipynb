```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame([
    {'speed': 14,
     'version': 'scikit-learn 0.20.1 (n_jobs=1)'},
    {'speed': 82,
     'version': 'vtext (n_jobs=1)'},
    {'speed': 303,
     'version': 'vtext (n_jobs=4)'}
]
)
```

```python
ax = df.set_index('version')[['speed']].plot.bar(
    legend=False,
    rot=10,
    width=0.4,
    alpha=0.7
)
ax.set(
    xlabel=None,
    title='CountVectorizer().transform speed',
    ylabel='Vectorization speed (MB/s)',
);
```
